```python
import pandas as pd
import requests
import math
from bs4 import BeautifulSoup

class Searcher:
    busca = {
        "uf": "SE",
        "oab": '353A',
        "de": '01/01/2023',
        "ate": '20/03/2024',
        "tipo": 'tiporpv'
    }
    data_dict = {
                'Processo': [],
                'Classe': [],
                'Data do Movimento': [],
                'Hora do Movimento': [],
                'Última Movimentação': []
            }
    page = None
    pages = 0
    contagem = 0

    def search(self) -> bool:
        self.link = f"https://cp.trf5.jus.br/processo/rpvprec/rpvPrecOAB/porData/{self.busca['tipo']}/ativos/{self.busca['uf']}000{self.busca['oab']}/{self.busca['de']}/{self.busca['ate']}//11"
        print(self.link)
        #self.link = f"https://cp.trf5.jus.br/processo/rpvprec/rpvPrecOAB/porData/{self.busca['tipo']}/ativos/{self.busca['uf']}000{self.busca['oab']}/{self.busca['de']}/{self.busca['ate']}//"
        site= self.nextPage()
        self.site = site
        self.pages = self.calcular_pages(site)
        
        self.extrair_dados(self.site)
        #for i in range(self.pages):
            #self.controller.atualizar_progresso(self.page, self.pages)
            #if i > 0:
                #site, self.page = self.nextPage()
        self.df_rpvs = pd.DataFrame.from_dict(self.data_dict)
        #self.contagem = self.df_rpvs.shape[0]

        return self.site

    def requesitar_dados (self, path):
        index = requests.get(path)
        return  BeautifulSoup(index.text, "html.parser")

    def calcular_pages(self, site):
        span = site.find(class_="texto_consulta")
        quantidade_resultados = int(span.text.split(" ")[1])
        result = math.ceil(quantidade_resultados / 10)
        return result

    def nextPage(self):
        if self.page == None:
            self.page = 0
        else:
            self.page = self.page + 1
        return self.requesitar_dados(self.link)

    def extrair_dados(self, site):
        tbody = site.find("tbody")
        trs = tbody.find_all('tr')
        for index in range(len(trs)):
            try:
                print(trs[index]['id'])
                tds = trs[index-1].find_all('td')
                self.data_dict['Processo'].append(tds[1].text)
                if len(tds) > 2:
                    self.data_dict['Classe'].append(tds[2].text)
                self.data_dict['Data do Movimento'].append(tds[3].text)
                self.data_dict['Hora do Movimento'].append(tds[4].text)
                self.data_dict['Última Movimentação'].append(tds[5].text)
            except KeyError:
                pass

    def export_to_excel(self, path):
        de_formatado = self.busca['de'].replace("/", "")
        ate_formatado = self.busca['ate'].replace("/", "")
        if self.busca['tipo'] == 'tiporpv':
            tipo = 'RPV'
        else:
            tipo = 'PREC'
        self.df_rpvs.to_excel(f"{path}\\{tipo}-{self.busca['oab']}-{de_formatado}-{ate_formatado}.xlsx")
```
